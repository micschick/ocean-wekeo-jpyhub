```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

LONS = np.arange(-180,180,5)
LATS = np.cos(LONS/360 * (2*np.pi))*90
SWH = 2*np.random.random(len(LONS)) - 1
SWH[SWH<0.000001] = 0.000001

GRADX = np.zeros(len(LONS))
GRADY = np.zeros(len(LONS))

for ii in np.arange(len(LONS)):
    if ii == 0 or ii == len(LONS) -1:
        continue
    else:
        GRADX[ii] = (LONS[ii+1] - LONS[ii-1])
        GRADY[ii] = (LATS[ii+1] - LATS[ii-1])

fig1 = plt.figure(figsize=(20, 20), dpi=300)
plt.plot(LONS,LATS,'k')
plt.quiver(LONS, LATS, -1*GRADY, GRADX,\
           color=plt.cm.Spectral_r(SWH/np.nanmax(SWH)),\
           scale=1/SWH*250, width=np.nanmean(GRADY)*0.5,headaxislength=0,headlength=0)
plt.quiver(LONS, LATS, GRADY, -1*GRADX,\
           color=plt.cm.Spectral_r(SWH/np.nanmax(SWH)),\
           scale=1/SWH*250, width=np.nanmean(GRADY)*0.5,headaxislength=0,headlength=0)
plt.axes().set_aspect('equal')
plt.show()
```
